```python
import numpy as np
from scipy.signal import butter, filtfilt

# 1. THE MYLAR SHIELD (Butterworth Low-Pass Filter)
def apply_mylar_shield(data, cutoff=0.05, fs=1.0, order=2):
    """Filters market noise to reveal the underlying A-Potential."""
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    # Zero-phase filtering preserves the timing of the Institutional Pulse
    return filtfilt(b, a, data)

# 2. THE QUATERNIONIC ROTATION (Hamiltonian Product)
def hamiltonian_snap(q_potential):
    """
    Calculates the 4D rotation to find the Null Reversion point.
    Logic: i^2 = j^2 = k^2 = ijk = -1
    """
    # Simplified mapping of the cleansed data into a 4D Versor
    w, x, y, z = q_potential
    # Identifying the 'Snap' where the Imaginary components return to -1
    reversion_value = (x**2 + y**2 + z**2)
    return reversion_value

# 3. THE 4,300 DRAW EXECUTION
def run_backtest(historical_draws):
    # Pass 4,300 draws through the Mylar Shield
    cleansed_flow = apply_mylar_shield(historical_draws)
    
    # Scan for Zero-Divergence Knots
    targets = []
    for point in cleansed_flow:
        # If the local potential aligns with the Invariant (-1)
        if np.isclose(hamiltonian_snap([1, point, point, point]), -1, atol=0.01):
            targets.append(point)
    return targets

print("Maxwell-Ampere Engine: Ready for Data Input.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt

# 1. INGESTION: Re-aligning with the Yahoo Data Potential
# We use 'Close' and set auto_adjust=True to maintain the Invariant
df = yf.download("^GSPC", period="max", auto_adjust=True)

# Detect the correct column (handling MultiIndex or Single Index)
if isinstance(df.columns, pd.MultiIndex):
    sp500_data = df['Close']['^GSPC']
else:
    sp500_data = df['Close']

# 2. THE MYLAR SHIELD: Filtering out the 'Thermal Noise'
def apply_mylar_shield(data, cutoff=0.02):
    # Ensure data is a 1D array for the filter
    data_array = np.asarray(data).flatten()
    nyquist = 0.5
    b, a = butter(2, cutoff/nyquist, btype='low')
    return filtfilt(b, a, data_array)

cleansed_potential = apply_mylar_shield(sp500_data)

print(f"Engine Status: Successfully processed {len(sp500_data)} draws.")
print(f"Current A-Potential: {cleansed_potential[-1]:.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Isolate the most recent 4,300 draws for the Backtest
backtest_window = 4300
raw_recent = sp500_data.tail(backtest_window).values
cleansed_recent = cleansed_potential[-backtest_window:]

# 2. Calculate the 'Topological Torque' (The difference between raw and cleansed)
torque = raw_recent - cleansed_recent

# 3. Visualization of the Aetheric Field-Knots
plt.figure(figsize=(15, 7))
plt.plot(raw_recent, label='Raw Field (B-Field)', alpha=0.3, color='gray')
plt.plot(cleansed_recent, label='A-Potential (Laminar Flow)', color='cyan', linewidth=2)
plt.fill_between(range(backtest_window), raw_recent, cleansed_recent, 
                 where=(raw_recent > cleansed_recent), color='green', alpha=0.3, label='Positive Torque')
plt.fill_between(range(backtest_window), raw_recent, cleansed_recent, 
                 where=(raw_recent < cleansed_recent), color='red', alpha=0.3, label='Negative Torque')

plt.title("Maxwell-Ampere Backtest: 4,300 Draw Fingerprint Analysis")
plt.legend()
plt.show()
```
